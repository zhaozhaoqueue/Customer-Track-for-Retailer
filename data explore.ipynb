```python
def cal_score(row):
    pid = row["ProductID"]
    price = prodDF.loc[prodDF["ProductID"]==pid, "UnitPrice"].values[0]
    cid = prodDF.loc[prodDF["ProductID"]==pid, "CategoryID"].values[0]
    score_level = catgDF.loc[catgDF["CategoryID"]==cid, "scoreLevel"]
    return row["Quantity"]*price*score_level
saleDF["score"] = saleDF.apply(cal_score, axis=1)
```

```python
custDF.insert(6, "Score", 0)
```

```python
def cal_cust_score(row):
    score = row["score"]
    cid = row["CustomerID"]
    custDF.loc[custDF["CustomerID"]==cid, "Score"] += score
saleDF.apply(cal_cust_score, axis=1)
```
